### Bibliothèques:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize':(15,8)})
```

### Import du data set:

```python
lifetime_earn = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/lifetime_earn.csv')
student_debt = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/student_debt.csv')
retirement = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/retirement.csv')
home_owner = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/home_owner.csv')
race_wealth = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/race_wealth.csv')
income_time = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/income_time.csv')
income_limits = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/income_limits.csv')
income_aggregate = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/income_aggregate.csv')
income_distribution = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/income_distribution.csv')
income_mean = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-02-09/income_mean.csv')
```

# 1 - Descriptions des différentes tables:

## lifetime_earn :

<ins>Taille</ins> : 6 lignes, 3 colonnes  
  
<ins>Description</ins> : Average lifetime earning by race/gender.  
  
<ins>Description des variables</ins> :  
gender	character	gender column  
race	character	Racial group  
lifetime_earn	double	Lifetime earnings

```python
lifetime_earn
```

## student_debt :

<ins>Taille</ins> : 30 lignes, 4 colonnes  
  
<ins>Description</ins> : Average family student loan debt for aged 25-55, by race and year normalized to 2016 dollars.  
  
<ins>Description des variables</ins> :  
year	double	Year of measure  
race	character	Racial group  
loan_debt	double	Loan debt  
loan_debt_pct	double	Share of families with student loan debt

```python
student_debt
```

## retirement :

<ins>Taille</ins> : 30 lignes, 3 colonnes  
  
<ins>Description</ins> : Average family liquid retirement savings normalized to 2016 dollars.  
  
<ins>Description des variables</ins> :  
year	double	Year  
race	character	Racial group  
retirement	double	Retirement dollars

```python
retirement
```

## home_owner :

<ins>Taille</ins> : 123 lignes, 3 colonnes  
  
<ins>Description</ins> : Home ownership percentage for families by race/year.  
  
<ins>Description des variables</ins> :  
year	double	Year  
race	character	Racial group  
home_owner_pct	double	Home ownership by race/ethnicity

```python
home_owner
```

## race_wealth :

<ins>Taille</ins> : 96 lignes, 4 colonnes  
  
<ins>Description</ins> : Family wealth by race/year/measure normalized to 2016, with measures of central tendency with mean and median.  
  
<ins>Description des variables</ins> :  
type	character	Type of measure, either median or mean  
year	double	Year  
race	character	Racial group  
wealth_family	double	Family wealth by race/year/measure normalized to 2016.

```python
race_wealth
```

## income_time :

<ins>Taille</ins> : 156 lignes, 3 colonnes  
  
<ins>Description</ins> : Family-level income by percentile and year.  
  
<ins>Description des variables</ins> :  
year	integer	Year  
percentile	character	Income percentile (10th, 50th, 90th)  
income_family	double	Familial income

```python
income_time
```

## income_limits :

<ins>Taille</ins> : 3090 lignes, 6 colonnes  
  
<ins>Description</ins> : Familial income limits for each fifth by year and race.  
  
<ins>Description des variables</ins> :  
year	double	Year  
race	character	Racial group  
dollar_type	character	Dollars in that year or normalized to 2019  
number	double	Number of households by racial group  
income_quintile	character	Income quintile as well as top 5%  
income_dollars	double	Income in US dollars, specific to dollar type  

```python
income_limits
```

## income_aggregate :

<ins>Taille</ins> : 1854 lignes, 5 colonnes  
  
<ins>Description</ins> : Share of aggregate income received by each fifth and top 5% of each racial group/household.  
  
<ins>Description des variables</ins> :  
year	double	Year  
race	character	Racial group  
number	double	Number of households by racial group  
income_quintile	character	Income quintile and/or top 5%  
income_share	double	Income share as a percentage  

```python
income_aggregate
```

## income_distribution :

<ins>Taille</ins> : 2916 lignes, 9 colonnes  
  
<ins>Description</ins> : Households by total money income, race, and hispanic origin of householder separated by year and income groups.  
  
<ins>Description des variables</ins> :  
year	integer	Year  
race	character	Racial Group  
number	double	Number of households  
income_median	integer	Income median  
income_med_moe	integer	Income median margin of error  
income_mean	integer	Income mean  
income_mean_moe	integer	Income mean margin of error  
income_bracket	character	Income bracket (9 total brackets between <$ 15,000 and >$ 200,000  
income_distribution	double	Income distribution as the percentage of each year/racial group - should add up to 100 for a specific year and race.

```python
income_distribution
```

## income_mean :

<ins>Taille</ins> : 3708 lignes, 5 colonnes  
  
<ins>Description</ins> : Mean income received by each fifth and top 5% of each racial group.  
  
<ins>Description des variables</ins> :  
year	double	Year  
race	character	Racial group  
dollar_type	character	Dollar type, ie dollar relative to that year or normalized to 2019  
income_quintile	character	Income quintile and/or top 5%  
income_dollars	double	Income dollar average

```python
income_mean
```

# 2 - Analyse exploratoire des données

```python
sns.lineplot(
      x=student_debt.year,
      y=student_debt.loan_debt,
      hue=student_debt.race,
      data=student_debt
)
```

```python
sns.lineplot(
      x=retirement.year,
      y=retirement.retirement,
      hue=retirement.race,
      data=retirement
)
```

```python
sns.lineplot(
      x=home_owner.year,
      y=home_owner.home_owner_pct,
      hue=home_owner.race,
      data=home_owner
)
```

```python
sns.lineplot(data=race_wealth, x=race_wealth.year, y=race_wealth.wealth_family, hue=race_wealth.race, style="type")
```

```python
sns.lineplot(x=income_time.year, y=income_time.income_family, hue=income_time.percentile, data=income_time)
```

```python
fig, axs = plt.subplots(ncols=3, nrows=2)

sns.lineplot(x='year', y='income_dollars', hue="race", data=income_limits.loc[(income_limits.income_quintile=="Lowest") & (income_limits.dollar_type=="2019 Dollars"),], ax=axs[0, 0])
sns.lineplot(x='year', y='income_dollars', hue="race", data=income_limits.loc[(income_limits.income_quintile=="Second") & (income_limits.dollar_type=="2019 Dollars"),], ax=axs[0, 1])
sns.lineplot(x='year', y='income_dollars', hue="race", data=income_limits.loc[(income_limits.income_quintile=="Third") & (income_limits.dollar_type=="2019 Dollars"),], ax=axs[0, 2])
sns.lineplot(x='year', y='income_dollars', hue="race", data=income_limits.loc[(income_limits.income_quintile=="Fourth") & (income_limits.dollar_type=="2019 Dollars"),], ax=axs[1, 0])
sns.lineplot(x='year', y='income_dollars', hue="race", data=income_limits.loc[(income_limits.income_quintile=="Top 5%") & (income_limits.dollar_type=="2019 Dollars"),], ax=axs[1, 1])
```

```python
sns.lineplot(x='year', y='number', hue="race", data=income_aggregate)
```

```python
fig, axs = plt.subplots(ncols=3, nrows=2)

sns.lineplot(x='year', y='income_share', hue="race", data=income_aggregate.loc[income_aggregate.income_quintile=="Lowest",], ax=axs[0, 0])
sns.lineplot(x='year', y='income_share', hue="race", data=income_aggregate.loc[income_aggregate.income_quintile=="Second",], ax=axs[0, 1])
sns.lineplot(x='year', y='income_share', hue="race", data=income_aggregate.loc[income_aggregate.income_quintile=="Third",], ax=axs[0, 2])
sns.lineplot(x='year', y='income_share', hue="race", data=income_aggregate.loc[income_aggregate.income_quintile=="Fourth",], ax=axs[1, 0])
sns.lineplot(x='year', y='income_share', hue="race", data=income_aggregate.loc[income_aggregate.income_quintile=="Highest",], ax=axs[1, 1])
sns.lineplot(x='year', y='income_share', hue="race", data=income_aggregate.loc[income_aggregate.income_quintile=="Top 5%",], ax=axs[1, 2])
```

## ACP Sur les variables quanti du jeu de données

```python
cls = PCA(n_components = 4)
pcs = cls.fit_transform(penguins.loc[:,("bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g")])
```

```python
cls.explained_variance_ratio_ #Presque 100% de l'information est présente sur l'axe 1
```

```python
pcs_Y = cls.transform(penguins.loc[:,("bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g")])
```

```python
df_pca = pd.DataFrame(pcs_Y, columns=["X1", "X2", "X3", "X4"])
df_pca = df_pca.join(penguins.loc[:,("species", "island", "sex", "year")])
df_pca
```

## CAH sur l'ACP

Critètes utilisables : ['ward', 'complete', 'average', 'single']

```python
from sklearn.cluster import AgglomerativeClustering
cls = AgglomerativeClustering(linkage="ward", affinity="euclidean", distance_threshold=0, n_clusters=None)
```

```python
cah = cls.fit(pcs_Y)
```

```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    default_kwargs = dict(leaf_font_size=10)
    default_kwargs.update(kwargs or {})

    dendrogram(linkage_matrix, **default_kwargs)
```

```python
plot_dendrogram(cah) # On retrouve surement les espèces
```
